```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
r = pd.read_csv('./data/R2014-2019.csv')
mean_r = r['Mean(R)']
zscore = ((mean_r-mean_r.mean())/mean_r.std())
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,5))
sns.distplot(mean_r,ax=ax[0])
ax[0].set_title('Density plot, R(t) Influneza',fontsize=15)
sns.distplot(zscore,ax=ax[1])
ax[1].set_title('Z-score of R for Influenza',fontsize=15)
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,5))
ax[0].plot(mean_r)
ax[0].set_title('R for Influenza by days',fontsize=15)
ax[1].plot(zscore)
ax[1].set_title('R (Z-score) for Influenza by days',fontsize=15)
```

```python
"""
Transforming z-score to R for COVID19
"""
rcovid =  zscore*mean_r.std()+2.2 # 100% SD
rcovid2 =  zscore*mean_r.std()*0.5+2.2 # 50% SD

fig, ax = plt.subplots(1,2,figsize=(20,5))
ax[0].plot(rcovid)
ax[0].set_title('R (100% SD of SD for R Influneza) for COVID19 by days',fontsize=15)
ax[1].plot(rcovid2)
ax[1].set_title('R (50% SD of SD for R Influneza) for COVID19 by days',fontsize=15)
ax[0].set_ylim(1,4.5)
ax[1].set_ylim(1,4.5)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.distplot(rcovid,ax=ax)
sns.distplot(rcovid2,ax=ax)
```
